The aim of this notebook is to design a funtion that locates the ground thuth texts aligned with Passim.

```python
import json
from pprint import pprint
```

```python
# first char and len of the GT line in the GT file
GT_start = 37637
GT_len = 34

# Load the index
GT_index_path = '4_finding_aligned_GT_references_files/index_Machzor_Yom_Kippur_Ashkenaz.json'
with open(GT_index_path, 'r') as file_handler:
    GT_index = json.load(file_handler)

# Finding the line in the index
for line in GT_index['lines']:
    if line['start_char'] >= GT_start:
        print(f"Found line start at: {line['start_char']} / line uid: {line['uid']}")
        print(f"Titles: {line['parent_titles']}")
        break
```
